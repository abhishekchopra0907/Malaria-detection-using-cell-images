```
import numpy as np
import math
import random as rd
import matplotlib.pyplot as plt
import math
from sklearn.preprocessing import MinMaxScaler

```

```
def sigmoid(num):
  a= 1.0/(1.0+np.exp(-1.0*num))
  return a
    
def relu(x):
  return np.maximum(x,0)

def tanh(x):
  return np.tanh(x)
```

```


def der_relu(z):
    return np.where(z <= 0, 0, 1)
def tanh_deriv(x):
    return 1.0 - np.tanh(x)**2
```

```


def linear_func(input_node,weight,bias):
  ##node_output=np.dot(input_node,weight)+bias
  node_output=np.dot(input_node,weight)
  
  return node_output/20
```

```
def bias_wieght_initial(input_dimention,output_node1,output_node2):
  weight1=np.random.rand(input_dimention,output_node1)
  weight2=np.random.rand(output_node1,output_node2)
  bias1=np.zeros(output_node1)
  bias2=np.zeros(output_node2)
  return weight1,weight2,bias1,bias2
```

```
def linear_block(input_to_node,weight,bias):
    node_output =np.dot(input_to_node,weight) +bias
    return node_output
```

```
def der_sigmoid(x):
  return x*(1-x)
##differential of sigmoid is sigmoid*(1-sigmoid)
```

```
def forward_pass(data,weight1,bias1,weight2,bias2):
  linear1=linear_func(data,weight1,bias1)
  ##we are having to do this element wise because sigmoid func only takes in an integar value and not a array one 
  non_linear1=sigmoid(linear1)
  linear2=linear_func(non_linear1,weight2,bias2)
  print(linear2[0:3])
  return linear2,non_linear1
```

```

def backward_pass(data,output,hidden_output,weight1,weight2,bias1,bias2,error):
  partial_derivative_beta=np.dot(hidden_output.T,error)
  partial_derivative_alpha_withoutdata=np.multiply(np.dot(error,weight2.T),der_sigmoid(hidden_output))
  partial_derivative_alpha=np.dot(data.T,partial_derivative_alpha_withoutdata)
  delta_bias2=np.sum(error,axis=0)
  delta_bias1=np.sum(partial_derivative_alpha_withoutdata,axis=0)
  learning_rate=0.1
  updated_weight1=weight1-learning_rate*partial_derivative_alpha
  updated_weight2=weight2-learning_rate*partial_derivative_beta
  updated_bias1=bias1-learning_rate*delta_bias1
  updated_bias2=bias2-learning_rate*delta_bias2
  return updated_weight1,updated_bias1,updated_weight2,updated_bias2
```

```

def train(data,labels,test_data,test_labels,weight1,weight2,bias1,bias2,epoch):
  Total_train_loss=[]
  Total_test_loss=[]
  Number_of_datapoints =data.shape[0]
  for i in range(epoch):
    output,hidden=forward_pass(data,weight1,bias1,weight2,bias2)
    error=output-labels
    #error = keras.utils.to_categorical(error, 10)
    Squared_error_loss = np.sum(0.5*(output-labels)**2)
    Mean_squared_error_loss = Squared_error_loss / Number_of_datapoints
    weight1,bias1,weight2,bias2=backward_pass(data,output,hidden,weight1,weight2,bias1,bias2,error)
    print("\n")
    print("weight 1")
    print( weight1)
    print("\n")
    print("weight 2")
    
    print(weight2)
    print("\n")
    print("bias 1")
    
    print(bias1)
    print("\n")
    print("bias 2")
    
    print(bias2)
    output_val,hidden_val=forward_pass(test_data,weight1,bias1,weight2,bias2)
    yes=np.argmax(output_val,axis=1)
    Test_loss = np.sum(0.5*(output_val-test_labels)**2)
    Test_loss/=Number_of_datapoints
    Total_train_loss=np.append(Total_train_loss,Mean_squared_error_loss)
    Total_test_loss=np.append(Total_test_loss,Test_loss)
  return Total_train_loss,Total_test_loss,weight1,weight2,bias1,bias2,yes
  
```

```
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop
import matplotlib.pyplot as plt


num_classes,num_nueron = 10,50
epoch = 1

# the data, shuffled and split between train and test sets
(x_train, y_train), (x_test, y_test) = mnist.load_data()
y_train = keras.utils.to_categorical(y_train, 10)
y_test =error = keras.utils.to_categorical(y_test, 10)
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
weight1,weight2,bias1,bias2=bias_wieght_initial(x_train.shape[1],num_nueron,num_classes)

print(train(x_train,y_train,x_test,y_test,weight1,weight2,bias1,bias2,epoch))
```
